# New one

```python
# USAGE:
# برنامه در این بخش باید حدودا تعداد 100 اگهی را بررسی و لینک و نام و قیمت
# آنهارا برای بررسی بیشتر جزویات در آینده ذخیره میکند

total = 220 # حداقل 100 تا اگهی پیدا کنه
i = 0 
row = []
driver = webdriver.Edge()
scroll = driver.execute_script("window.scrollTo(0, document.body.scrollHeight)") # اسکرول انیجاس فقط برای زیبایی ساختار حلقه
driver.get(url)
driver.maximize_window()
while i <= total:
    i+=1
    try:
        article = driver.find_element(By.CSS_SELECTOR, f"""#app > div.kt-col-md-12-d59e3.browse-c7458 > main > div:nth-child(3) > div > 
                                      div > div > div:nth-child({i})""") # اون خونه از جدولی که کل دیتای اگهی رو داراس
        link = article.find_element(By.CSS_SELECTOR, f"""div:nth-child({i}) > a""").get_attribute('href')
        title = article.find_element(By.TAG_NAME, "h2").text
        price = article.find_element(By.XPATH, """a/article/div/div[1]/div[1]""").text
        ram = price.replace("تومان", "") # این خط و خط پایینی موقتا برای خالص سازی قیمت جهت فیلتر کردن آن در آینده
        ram = ram.replace(",", "")   
        price = int(digits.fa_to_en(ram)) 
        row.append((title, price, link))

    except Exception:
        #فقط زمانی برنامه وارد این قسمت خواهد شد که به انتهای صفحه رسیده و برای ادامه باید یک مقدار اسکرول کرد
        if i >= 14:
            total -= (i-1)
            scroll
            i = 4

column = ["Title", "Price", "Link"]
df = pd.DataFrame(row, columns=column)
df.to_csv("All_List(1).csv")
df
```

# Find Mean Price

```python
# Usage:
# برنامه در این بخش فایل از ذخیره شده در قسمت (New One) میانگین میگیرد
df = pd.read_csv("All_List(1).csv" )   
medium = int(df["Price"].mean())   
medium
```

# Filter By Prices 

```python
data = pd.read_csv("All_List(1).csv")
data.columns = ["unnamed", "Title", "Price", "Link"]
link_list = list(data.Link)
title_list = list(data.Title)
price_list = list(data.Price)
row = []
i = 0
for p in price_list:
    if p < medium:
        row.append((title_list[i], price_list[i], link_list[i]))
    i+=1
column = ["Title", "Price", "URL"]
df = pd.DataFrame(row, columns=column)
df.to_csv("Price_Filter(1).csv")
df
```

# Filter By Year

```python
data = pd.read_csv("Price_Filter(1).csv")
data.columns = ["unnamed", "Title", "Price", "URL"]
link_list = list(data.URL)
title_list = list(data.Title)
price_list = list(data.Price)
row = []
i = 0
for link in link_list:
    driver.get(link)
    built_year = driver.find_element(By.XPATH, """//*[@id="app"]/div[2]/div/div[1]/div[1]/div[4]/div[1]/div[2]/span[2]""").text
    built_year = int(digits.fa_to_en(built_year)) # سال ساخت
    if built_year >= 1395:
        meterage = driver.find_element(By.XPATH, """//*[@id="app"]/div[2]/div/div[1]/div[1]/div[4]/div[1]/div[1]/span[2]""").text
        meterage = int(digits.fa_to_en(meterage))
        room = driver.find_element(By.XPATH, """//*[@id="app"]/div[2]/div/div[1]/div[1]/div[4]/div[1]/div[3]/span[2]""").text
        room = int(digits.fa_to_en(room))
        row.append((title_list[i], price_list[i], meterage, built_year, room, link))
    i += 1

column = ["Title", "Price", "Meterage", "Built Year", "Rooms", "URL"]
df = pd.DataFrame(row, columns=column)
df.to_csv("Year_Filter(1).csv")
df
```
